```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read in the cleaned data
spotify = pd.read_csv('csv_outputs/cleaned_spotify.csv')

# Select only numeric columns
numeric_spotify = spotify.select_dtypes(include=[np.number])

# Correlation matrix
spotify_corr = numeric_spotify.corr()

spotify_corr.to_csv('csv_outputs/correlation_matrix.csv')
```

```python
# Making scatterplots for each variable vs popularity
columns = spotify.columns
for i in columns:
    for j in columns:
        if '64' in str(spotify[i].dtype) and '64' in str(spotify[j].dtype) and i != j and 'Unnamed' not in i and 'Unnamed' not in j:
            plt.plot(np.unique(spotify[i]), np.poly1d(np.polyfit(spotify[i], spotify[j], 1))(np.unique(spotify[i])), color='red')
            plt.scatter(spotify[i], spotify[j], s=5)
            plt.xlabel(i)
            plt.ylabel(j)
            plt.show()
```

```python
# Making correlation heatmap
plt.figure(figsize=(10, 8))  # Optional: Adjust the figure size
sns.heatmap(spotify_corr, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)

# Show the plot
plt.show()
```
